```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Reinitialize data
billing_codes = {
    "99453": 18.95,  # RPM Setup (one-time)
    "99454": 45.12,  # RPM Device Supply (monthly)
    "99457": 46.85,  # RPM Management (first 20 mins, monthly)
    "99458": 37.62,  # Additional RPM Management (monthly)
    "99490": 59.90,  # CCM by Clinical Staff (monthly)
    "99491": 80.98,  # CCM by Physician (1 in 20 patients, monthly)
    "99091": 51.29,  # Data Interpretation (monthly)
    "99495": 230.00  # TCM (one-time)
}

theoretical_billing_codes = {
    **billing_codes,
    "Alzheimer’s Prevention": 180.00,  # Hypothetical billing code
    "Mental Health Support": 120.00,   # Hypothetical billing code
    "Preventive Care": 200.00          # Hypothetical billing code
}

gpci_adjustments = {
    "Virginia": 1.0,
    "Florida": 1.05,
    "Texas": 1.03,
    "New York": 1.08,
    "New Jersey": 1.06,
    "California": 1.1
}

recurring_costs = {
    "Platform Fee": 5.00,
    "Staffing": 25.00
}

one_time_costs = {
    "Device": 185.00
}

# Function to calculate unit economics
def calculate_unit_economics(state, billing_codes, recurring_costs, one_time_costs, patients, theoretical=False):
    gpci = gpci_adjustments[state]
    codes = theoretical_billing_codes if theoretical else billing_codes
    
    recurring_revenue = (
        codes["99454"] + 
        codes["99457"] + 
        codes["99458"] + 
        codes["99490"] + 
        (codes["99491"] / 20) + 
        codes["99091"]
    )
    if theoretical:
        recurring_revenue += (
            codes["Alzheimer’s Prevention"] +
            codes["Mental Health Support"] +
            codes["Preventive Care"]
        )
    recurring_revenue *= gpci

    one_time_revenue = (codes["99453"] + codes["99495"]) * gpci

    total_recurring_cost = recurring_costs["Platform Fee"] + recurring_costs["Staffing"]
    total_one_time_cost = one_time_costs["Device"]

    recurring_ebitda = recurring_revenue - total_recurring_cost
    one_time_ebitda = one_time_revenue - total_one_time_cost

    return {
        "State": state,
        "Recurring Revenue": recurring_revenue,
        "Recurring Costs": total_recurring_cost,
        "Recurring EBITDA": recurring_ebitda,
        "One-Time Revenue": one_time_revenue,
        "One-Time Costs": total_one_time_cost,
        "One-Time EBITDA": one_time_ebitda
    }

patients = 150
unit_economics_realistic = [calculate_unit_economics(state, billing_codes, recurring_costs, one_time_costs, patients) for state in gpci_adjustments.keys()]
unit_economics_theoretical = [calculate_unit_economics(state, billing_codes, recurring_costs, one_time_costs, patients, theoretical=True) for state in gpci_adjustments.keys()]

unit_economics_realistic_df = pd.DataFrame(unit_economics_realistic)
unit_economics_theoretical_df = pd.DataFrame(unit_economics_theoretical)

# Enhanced Visualization
states = unit_economics_realistic_df["State"]
x = np.arange(len(states))
width = 0.4

custom_palette = [
    mcolors.CSS4_COLORS["mediumseagreen"],
    mcolors.CSS4_COLORS["indianred"],
    mcolors.CSS4_COLORS["cornflowerblue"],
    mcolors.CSS4_COLORS["darkorange"],
    mcolors.CSS4_COLORS["black"]
]

fig, ax = plt.subplots(2, 1, figsize=(14, 14))

# Realistic Unit Economics
ax[0].bar(x - width / 2, unit_economics_realistic_df["Recurring Revenue"], width,
          label='Recurring Revenue', color=custom_palette[0], alpha=0.8)
ax[0].bar(x - width / 2, -unit_economics_realistic_df["Recurring Costs"], width,
          label='Recurring Costs', color=custom_palette[1], alpha=0.8, hatch="//")
ax[0].bar(x + width / 2, unit_economics_realistic_df["One-Time Revenue"], width,
          label='One-Time Revenue', color=custom_palette[2], alpha=0.8)
ax[0].bar(x + width / 2, -unit_economics_realistic_df["One-Time Costs"], width,
          label='One-Time Costs', color=custom_palette[3], alpha=0.8, hatch="\\\\")
ax[0].plot(x, unit_economics_realistic_df["Recurring EBITDA"], label="Recurring EBITDA",
           color=custom_palette[4], marker='o', linestyle='--', linewidth=2)
ax[0].set_title("Realistic Unit Economics per Patient by State", fontsize=14, weight='bold')

# Theoretical Unit Economics
ax[1].bar(x - width / 2, unit_economics_theoretical_df["Recurring Revenue"], width,
          label='Recurring Revenue', color=custom_palette[0], alpha=0.8)
ax[1].bar(x - width / 2, -unit_economics_theoretical_df["Recurring Costs"], width,
          label='Recurring Costs', color=custom_palette[1], alpha=0.8, hatch="//")
ax[1].bar(x + width / 2, unit_economics_theoretical_df["One-Time Revenue"], width,
          label='One-Time Revenue', color=custom_palette[2], alpha=0.8)
ax[1].bar(x + width / 2, -unit_economics_theoretical_df["One-Time Costs"], width,
          label='One-Time Costs', color=custom_palette[3], alpha=0.8, hatch="\\\\")
ax[1].plot(x, unit_economics_theoretical_df["Recurring EBITDA"], label="Recurring EBITDA",
           color=custom_palette[4], marker='o', linestyle='--', linewidth=2)
ax[1].set_title("Theoretical Unit Economics per Patient by State", fontsize=14, weight='bold')

# Formatting
for axis in ax:
    axis.axhline(0, color='black', linewidth=0.8, linestyle='--')
    axis.set_xticks(x)
    axis.set_xticklabels(states, fontsize=10, weight='bold')
    axis.set_xlabel("States", fontsize=12, weight='bold')
    axis.set_ylabel("USD ($)", fontsize=12, weight='bold')
    axis.legend(loc="upper right", fontsize=10)
    axis.grid(True, linestyle=":", linewidth=0.6, alpha=0.7)
    for bar_group in axis.containers:
        axis.bar_label(bar_group, fmt="%.1f", fontsize=8, padding=3)

plt.tight_layout()
plt.show()
```
