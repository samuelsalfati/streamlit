```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Constants and initial parameters
states = ["Virginia", "Florida", "Texas", "New York", "California"]
gpci_adjustments = {"Virginia": 1.0, "Florida": 1.05, "Texas": 1.03, "New York": 1.08, "California": 1.1}
market_size_per_state = {"Virginia": 50000, "Florida": 70000, "Texas": 80000, "New York": 60000, "California": 90000}
nursing_homes_per_state = {"Virginia": 40, "Florida": 50, "Texas": 60, "New York": 30, "California": 70}
home_growth_rate = 0.05  # 5% annual growth in nursing homes
patients_per_home = 20  # Initial patients per nursing home
initial_cash = 1_500_000 - sum([500_000, 400_000, 300_000, 300_000])  # Remaining budget after initial spend
months = 60  # Simulation time in months

# Revenue and cost definitions
recurring_revenue_per_patient = {
    "99454": 45.12,
    "99457": 46.85,
    "99458": 37.62,
    "99490": 59.90,
    "99491": 80.98 / 20,
    "99091": 51.29,
}
non_recurring_revenue_per_patient = {"99453": 18.95, "TCM": 230.00}
theoretical_billing_codes = {
    "Alzheimer's Monitoring": 180.00,
    "Mental Health Support": 120.00,
    "Preventive Care": 200.00,
}
staffing_cost_per_patient = 50.00
platform_fee_per_patient = 5.00
hardware_cost_per_patient = 185.00
nurse_cost = 70_000 / 12 / 50
md_cost = 120_000 / 12 / (50 * 50)
operational_overhead_base = 10_000

# Initialize financial tracking
cash_balance = initial_cash
total_patients = {state: 0 for state in states}
start_month_for_state = {"Virginia": 1, "Florida": 13, "Texas": 13, "New York": 25, "California": 37}

# Data storage for results
data = {
    "Month": [],
    "State": [],
    "New Patients": [],
    "Total Patients": [],
    "Market Penetration (%)": [],
    "Recurring Revenue": [],
    "Non-Recurring Revenue": [],
    "Theoretical Revenue": [],
    "Total Revenue": [],
    "Recurring Costs": [],
    "Non-Recurring Costs": [],
    "Total Costs": [],
    "Actual EBITDA": [],
    "Theoretical EBITDA": [],
    "Cash Balance": [],
}

# Helper functions
def calculate_nursing_homes(state, month):
    """Calculate the number of nursing homes in a state over time."""
    return nursing_homes_per_state[state] * (1 + home_growth_rate) ** ((month - 1) // 12)

def calculate_patients(state, month, year_target, max_market_share):
    """Gradual onboarding of patients to align with market penetration goals."""
    growth_rate = max_market_share / year_target / 12  # Monthly growth rate for gradual scaling
    return int(total_patients[state] + growth_rate * market_size_per_state[state])

# Simulation
for month in range(1, months + 1):
    active_states = [state for state, start_month in start_month_for_state.items() if month >= start_month]

    for state in active_states:
        gpci = gpci_adjustments[state]
        nursing_homes = calculate_nursing_homes(state, month)
        max_patients = int(nursing_homes * patients_per_home)
        
        # Determine new patients this month
        year = (month - 1) // 12 + 1
        year_target = 3 if year <= 3 else 5  # Year 3 for 15%, Year 5 for 20%
        max_market_share = 0.15 if year <= 3 else 0.20  # Gradual growth target
        new_patients = calculate_patients(state, month, year_target, max_market_share) - total_patients[state]
        
        # Update total patients and ensure it doesn't exceed market size
        total_patients[state] += new_patients
        total_patients[state] = min(total_patients[state], market_size_per_state[state])

        # Revenue calculations
        recurring_revenue = sum(recurring_revenue_per_patient.values()) * gpci * total_patients[state]
        non_recurring_revenue = sum(non_recurring_revenue_per_patient.values()) * gpci * new_patients
        theoretical_revenue = sum(theoretical_billing_codes.values()) * gpci * total_patients[state]
        total_revenue = recurring_revenue + non_recurring_revenue

        # Cost calculations
        recurring_cost = (staffing_cost_per_patient + platform_fee_per_patient) * total_patients[state]
        nurse_md_cost = (nurse_cost + md_cost) * total_patients[state]
        non_recurring_cost = hardware_cost_per_patient * new_patients
        operational_overhead = operational_overhead_base + len(active_states) * 1_000
        operational_overhead = min(operational_overhead, 50_000)

        total_costs = recurring_cost + nurse_md_cost + non_recurring_cost + operational_overhead

        # EBITDA calculations
        actual_ebitda = total_revenue - total_costs
        theoretical_ebitda = total_revenue + theoretical_revenue - total_costs
        cash_balance += actual_ebitda

        # Record results
        data["Month"].append(month)
        data["State"].append(state)
        data["New Patients"].append(new_patients)
        data["Total Patients"].append(total_patients[state])
        data["Market Penetration (%)"].append((total_patients[state] / market_size_per_state[state]) * 100)
        data["Recurring Revenue"].append(recurring_revenue)
        data["Non-Recurring Revenue"].append(non_recurring_revenue)
        data["Theoretical Revenue"].append(theoretical_revenue)
        data["Total Revenue"].append(total_revenue)
        data["Recurring Costs"].append(recurring_cost)
        data["Non-Recurring Costs"].append(non_recurring_cost)
        data["Total Costs"].append(total_costs)
        data["Actual EBITDA"].append(actual_ebitda)
        data["Theoretical EBITDA"].append(theoretical_ebitda)
        data["Cash Balance"].append(cash_balance)

# Convert data to a DataFrame
financial_df = pd.DataFrame(data)

# Summarize and validate results
yearly_summary = financial_df.groupby(["State", financial_df["Month"] // 12 + 1]).agg(
    Market_Penetration=("Market Penetration (%)", "max"),
    Total_Patients=("Total Patients", "max"),
).reset_index()

# Restore Graphing
plt.figure(figsize=(12, 6))
for state in states:
    state_data = financial_df[financial_df["State"] == state].groupby("Month")["Total Revenue"].sum()
    plt.plot(state_data.index, state_data.values / 1_000_000, label=f"{state} Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue (in Millions)")
plt.title("Monthly Revenue by State")
plt.legend()
plt.grid(True)
plt.show()

# Theoretical vs Actual EBITDA
plt.figure(figsize=(12, 6))
financial_df.groupby("Month")[["Actual EBITDA", "Theoretical EBITDA"]].sum().plot()
plt.xlabel("Month")
plt.ylabel("EBITDA (in Millions)")
plt.title("Theoretical vs Actual EBITDA")
plt.grid(True)
plt.show()

# Display results to the user
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Constants and assumptions
states = ["Virginia", "Florida", "Texas", "New York", "California"]
gpci_adjustments = {
    "Virginia": 1.0,
    "Florida": 1.05,
    "Texas": 1.03,
    "New York": 1.08,
    "California": 1.1,
}
market_size_per_state = {
    "Virginia": 50000,
    "Florida": 70000,
    "Texas": 80000,
    "New York": 60000,
    "California": 90000,
}
nursing_homes_per_state = {
    "Virginia": 40,
    "Florida": 50,
    "Texas": 60,
    "New York": 30,
    "California": 70,
}

# Revenue and costs assumptions
recurring_revenue_per_patient = {
    "99454": 45.12,
    "99457": 46.85,
    "99458": 37.62,
    "99490": 59.90,
    "99491": 80.98 / 20,
    "99091": 51.29,
}
non_recurring_revenue_per_patient = {
    "99453": 18.95,
    "TCM": 230.00,
}
staffing_cost_per_patient = 50.00
platform_fee_per_patient = 5.00
hardware_cost_per_patient = 185.00
nurse_cost = 70_000 / 12 / 50
md_cost = 120_000 / 12 / (50 * 50)
operational_overhead_base = 10_000

# Initialization
initial_cash = 1_500_000 - sum([500_000, 400_000, 300_000, 300_000])
months = 60
cash_balance = initial_cash
data = {
    "Month": [],
    "State": [],
    "New Patients": [],
    "Total Patients": [],
    "Nursing Homes": [],
    "Market Penetration": [],
    "Recurring Revenue": [],
    "Non-Recurring Revenue": [],
    "Total Revenue": [],
    "Recurring Costs": [],
    "Non-Recurring Costs": [],
    "Total Costs": [],
    "Actual EBITDA": [],
    "Theoretical EBITDA": [],
    "Cash Balance": [],
}

# Helper functions
def calculate_new_patients(state, month, total_patients, max_patients):
    if month == 1:
        return 50
    elif total_patients >= max_patients:
        return 0
    else:
        return int(0.1 * total_patients + (max_patients - total_patients) * 0.05)

def calculate_nursing_homes(state, month):
    initial_homes = nursing_homes_per_state[state]
    growth_years = (month - 1) // 12
    return initial_homes + growth_years

# Financial model logic
for month in range(1, months + 1):
    for state in states:
        gpci = gpci_adjustments[state]
        nursing_homes = calculate_nursing_homes(state, month)
        max_patients = market_size_per_state[state]

        # Calculate new patients and update total patients
        total_patients_in_state = sum(
            data["New Patients"][-len(states):] if data["State"][-len(states):] == state else []
        )
        new_patients = calculate_new_patients(state, month, total_patients_in_state, max_patients)

        # Revenue calculations
        recurring_revenue = sum(recurring_revenue_per_patient.values()) * gpci * total_patients_in_state
        non_recurring_revenue = sum(non_recurring_revenue_per_patient.values()) * gpci * new_patients
        total_revenue = recurring_revenue + non_recurring_revenue

        # Cost calculations
        recurring_cost = (staffing_cost_per_patient + platform_fee_per_patient) * total_patients_in_state
        nurse_md_cost = (nurse_cost + md_cost) * total_patients_in_state
        non_recurring_cost = hardware_cost_per_patient * new_patients
        operational_overhead = min(
            operational_overhead_base + len(states) * 1_000, 50_000
        )
        total_costs = recurring_cost + nurse_md_cost + non_recurring_cost + operational_overhead

        # EBITDA calculations
        actual_ebitda = total_revenue - total_costs
        theoretical_ebitda = actual_ebitda + (0.15 * total_revenue)  # Assuming theoretical adds 15%
        cash_balance += actual_ebitda

        # Store results
        data["Month"].append(month)
        data["State"].append(state)
        data["New Patients"].append(new_patients)
        data["Total Patients"].append(total_patients_in_state)
        data["Nursing Homes"].append(nursing_homes)
        data["Market Penetration"].append(total_patients_in_state / max_patients)
        data["Recurring Revenue"].append(recurring_revenue)
        data["Non-Recurring Revenue"].append(non_recurring_revenue)
        data["Total Revenue"].append(total_revenue)
        data["Recurring Costs"].append(recurring_cost)
        data["Non-Recurring Costs"].append(non_recurring_cost)
        data["Total Costs"].append(total_costs)
        data["Actual EBITDA"].append(actual_ebitda)
        data["Theoretical EBITDA"].append(theoretical_ebitda)
        data["Cash Balance"].append(cash_balance)

# Create DataFrame
financial_df = pd.DataFrame(data)
financial_df["Year"] = (financial_df["Month"] - 1) // 12 + 1

# Summarize data by year and state
summary = financial_df.groupby(["Year", "State"]).agg(
    Annual_Revenue=("Total Revenue", "sum"),
    Annual_EBITDA=("Actual EBITDA", "sum"),
    Annual_Theoretical_EBITDA=("Theoretical EBITDA", "sum"),
).reset_index()

# Plot Revenue by State
plt.figure(figsize=(12, 6))
for state in states:
    state_data = financial_df[financial_df["State"] == state].groupby("Month")["Total Revenue"].sum()
    plt.plot(state_data.index, state_data.values / 1_000_000, label=f"{state} Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue (in Millions)")
plt.title("Monthly Revenue by State")
plt.legend()
plt.grid(True)
plt.show()

# Plot Actual and Theoretical EBITDA by State
fig, ax1 = plt.subplots(figsize=(12, 6))
for state in states:
    state_data = financial_df[financial_df["State"] == state].groupby("Month")["Actual EBITDA"].sum()
    ax1.bar(
        state_data.index, state_data.values / 1_000_000, label=f"{state} EBITDA", alpha=0.7
    )

# Theoretical EBITDA Overlay
ax2 = ax1.twinx()
ax2.plot(
    financial_df.groupby("Month")["Theoretical EBITDA"].sum().index,
    financial_df.groupby("Month")["Theoretical EBITDA"].sum().values / 1_000_000,
    label="Theoretical EBITDA",
    color="red",
    linewidth=2,
)

# Labels and Formatting
ax1.set_xlabel("Month")
ax1.set_ylabel("Actual EBITDA (in Millions)")
ax2.set_ylabel("Theoretical EBITDA (in Millions)")
ax1.set_title("Actual and Theoretical EBITDA")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
plt.grid(True)
plt.show()

# Display Summary Table
print(summary)
```
